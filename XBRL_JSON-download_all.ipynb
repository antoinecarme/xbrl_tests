```python
import json

with open('xbrl_index.json') as f:
    gXBRL_JSON = json.load(f)
```

```python
lJSONFiles = {}
lCountries = {}
lURI = "https://filings.xbrl.org/"
for k in gXBRL_JSON.keys():
    lFilings = gXBRL_JSON[k]['filings']
    for (k1, v1) in lFilings.items():
        lXBRL_JSON = v1.get('xbrl-json')
        if(lXBRL_JSON):
            lCountry = v1.get('country')
            x = "data/" + lCountry + "/"
            lCountries[k] = (x, "mv \"data/" + lXBRL_JSON + "\" \"" + x + lXBRL_JSON + "\" ")
            # print((k , k1, lXBRL_JSON, k + "/" + lXBRL_JSON))
            lJSONFiles[k] = "wget --quiet -O \"data/" + lXBRL_JSON + "\" " + '"' + lURI + "/" + k1 + "/" + lXBRL_JSON + '"'
```

```python
import os
for (k, v) in [x for x in lJSONFiles.items()][:10]:
    print("RETRIEVING" , k, v)
    # os.system(v)
```
