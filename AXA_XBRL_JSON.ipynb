```python
def pretty_print(iFact):
    print(iFact)
    
def dump_fact(iFactIdx):
    lFact = gFacts[list(gFacts.keys())[iFactIdx]]
    pretty_print(lFact)
```

```python
dump_fact(0)
```

```python
dump_fact(gNBFacts - 1)
```

```python
def add_dimensions():
    lDict = {}
    for lFactIdx in range(gNBFacts):
        lFact = gFacts[list(gFacts.keys())[lFactIdx]]
        # print(lFact)
        # print(str(lFact.get('value', ''))[:30], lFact['dimensions'])
        lDict2 = {}
        for (k, v) in lFact['dimensions'].items():
            lDict2[k] = v
        lDict2['value'] = str(lFact.get('value', ''))[:30]
        lConcept = lDict2.get('concept')
        lDict[lConcept] = lDict.get(lConcept , {})
        lDict[lConcept][lFactIdx] = lDict2
    return lDict
```

```python
gFacts2 = add_dimensions()
```

```python
import numpy as np
import pandas as pd
lDataFrameByConcept = {}
for (lConcept, lDict) in gFacts2.items():
    lDataFrameByConcept[lConcept] = pd.DataFrame.from_dict(lDict, orient='index')
    display(lDataFrameByConcept[lConcept])
```

```python
for (lConcept, lDF) in lDataFrameByConcept.items():
    if(lDF.shape[0] > 5):
        print("###### Concept info " + lConcept + " " + str(lDF.shape) + " ####################")
        print(lConcept, lDF.columns)
        for col in lDF.columns:
            print((lConcept , col, lDF[col].value_counts().to_dict()))
```
